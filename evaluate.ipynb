```python
import pandas as pd
import pypsa
import glob
```

```python
capacity_table = pd.DataFrame()

for ct in glob.glob("resources/capacity_tables/*.csv"):
    iteration = int(ct.split(".")[0].split("tables/")[1])
    capacity_table[iteration] = pd.read_csv(ct, index_col=[0,1]).stack()
```

```python
capacity_table.loc[:, :, "p_nom"].sum().sort_index().plot()
```

```python
capacity_table.filter(like="UK",axis=0).loc[:, 2025, "p_nom"].sum().sort_index().plot()
```
